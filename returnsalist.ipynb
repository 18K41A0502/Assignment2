#14.  Write a Python code that takes a number and returns a list of its digits. So for 586392 it should return [5,8,6,3,9,2].

```
def return_list_of_digits(number_1):
    c = str(number_1)
    int_list = []
    for i in range(len(c)):
        int_list.append(int(c[i]))
    return int_list
input_number = input("Enter the number: ")
list_of_digits = return_list_of_digits(input_number)
print(list_of_digits)
```
